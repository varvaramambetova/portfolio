### Подготовим данные

####  Посмотрим, как менялись продажи по платформам. Платформы с наибольшими суммарными продажами , распределение по годам

```python
data.groupby(by='platform')['sum_sales'].sum().sort_values(ascending=False).plot(
kind='bar',
ylabel = 'sum',
figsize=(10, 5));
plt.title('Продажи по платформам')
plt.xlabel('Платформы')
plt.ylabel('Количество продаж, миллионы проданных копий')
plt.show()
```

Платформы с самыми высокими продажами за всё время PS2, X360,PS3, Wii, DS, PS, PS4, GBA, PSP, 3DS, PC,GB
спорный лидер продаж - PS2. Остальные платформы продали намного меньше копий

```python
top =['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS', 'PS4', 'GBA', 'PSP', '3DS','PC', 'GB']
for platform in top:
    data[data['platform'] == platform].pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum').plot(
    kind='bar',
    figsize=(10,5))
    plt.title(f'Продажи на платформе {platform} по годам')
    plt.xlabel('года продаж')
    plt.ylabel('Количество продаж, миллионы проданных копий')
    plt.show()
```

По гарфикам видно, что новые платформы появляются каждые 10 лет, и через 10 лет исчезают. Первая пятерка по количеству продаж в данный период уже не актуально и наблюдается снижение продаж . Использовать хти данные нельзя

#### Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков, построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами

```python
ps4_corr_critic_score=data_after_2014.query('platform == "PS4" & critic_score != -1')

ps4_corr_user_score=data_after_2014.query('platform == "PS4" & user_score!= -1')
```

```python
top_after_2014 =['PS4', 'XOne', '3DS', 'PS3','X360', 'WiiU', 'PC', 'PSV', 'Wii', 'PSP']
for platform in top_after_2014:
    data_after_2014[data_after_2014['platform'] == platform].pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum').plot(
    kind='bar',
    figsize=(10,5))
    plt.title(f'Продажи на платформе {platform} c 2014 года')
    plt.xlabel('года продаж')
    plt.ylabel('Количество продаж, миллионы проданных копий')
    plt.show()
```

```python
ps4_corr_critic_score.plot(x='critic_score', y='sum_sales', kind='scatter',  grid=True, alpha=0.3 ,figsize=(10, 5)) 
plt.title('Диаграмма рассеяния зависимости продаж PS4 от оценки критиков')
plt.xlabel('Оценка критика')
plt.ylabel('Сумарные продажи')
plt.show()
```

```python
#Проведем расчёт корреляции между оценкой критиков и продажами
ps4_corr_critic_score['critic_score'].corr(ps4_corr_critic_score['sum_sales'])
```

```python
ps4_corr_user_score.plot(x='user_score', y='sum_sales', kind='scatter',  grid=True, alpha=0.3 ,figsize=(10, 5)) 
plt.title('Диаграмма рассеяния зависимости продаж PS4 от оценки пользователей')
plt.xlabel('Оценка пользователя')
plt.ylabel('Сумарные продажи')
plt.show()
```

```python
#Проведем расчёт корреляции между оценкой пользователей и продажами
ps4_corr_user_score['user_score'].corr(ps4_corr_user_score['sum_sales'])                         
```

По первом графике видно, что оценка критиков влияет на продажи, но не во всех случаях, например, есть на графике точки с выскоими оценками, но с худшими продажами, чему некоторых игр с меньшей оценкой. При расчетах коэффициент Пирсона показывает умеренную корреляцию
На втором графике видно, что игры с маленькими оценками чаще выходят на высокие продажи, игры с самыми высокими оценками пользователей редко дают высокие продажи, на графике видно лишь одно выбивающееся значение, где сошлась высокая пользовательская оценка и высокие продажи. При рассчетах коэффициент Пирсона показывает что связи между пользовательской оценкой и продажами.

### Шаг 5. Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;


H₀: средние пользовательские рейтинги платформы Xbox One = PC средние пользовательские рейтинги платформы Xbox One.  
H₁: средние пользовательские рейтинги платформы Xbox One != PC средние пользовательские рейтинги платформы Xbox One.  
alpha = 0.05

```python
sample_1 = data_after_2014.query('platform == "XOne" & user_score != -1')
sample_2 = data_after_2014.query('platform == "PC" & user_score != -1')
alpha = .05 

results = st.ttest_ind(sample_1['user_score'], sample_2['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
sample_1.pivot_table(index = 'platform', values ='user_score', aggfunc='mean' )
```

```python
sample_2.pivot_table(index = 'platform', values ='user_score', aggfunc='mean' )
```

Вывод: Поскольку pvalue не меньше уровня значимости α = 0,05, мы не можем отвергнуть нулевую гипотезу.   
У нас нет достаточных доказательств, чтобы сказать, что средние пользовательские рейтинги между этими двумя платформами различен.

#### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

H₀: средние пользовательские рейтинги жанров Action =  средние пользовательские рейтинги жанра Sports  

H₁: средние пользовательские рейтинги жанров Action != средние пользовательские рейтинги жанра Sports
alpha = 0.05

```python
samp_1 = data_after_2014.query('genre == "Action" & user_score != -1')
samp_2 = data_after_2014.query('genre == "Sports" & user_score != -1')
alpha = 0.05 

results = st.ttest_ind(samp_1['user_score'], samp_2['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Вывод: Поскольку pvalue не больше уровня значимости α = 0,05, мы можем отвергнуть нулевую гипотезу.   
Вероятность мала, получить одинаковые средние пользовательские рейтинги между этими двумя жанрами.
