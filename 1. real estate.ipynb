### Проведем предобработку данных

#### Переименование столбцов

```python
data = data.rename(columns={'parks_around3000': 'parks_around_3000','cityCenters_nearest':'сity_centers_nearest', 'ponds_around3000': 'ponds_around_3000'}) 

#Проверим изменились названия столбцов
print(data.columns)
```

```python
data.info()
```

#### Обработка пропусков

```python
data.isna().sum()
```

##### Замена пропусков

```python
#Пропуски в данных в locality_name заменяем на 'Другие'
data['locality_name'] = data['locality_name'].fillna('Другие')
#Пропуски в ceiling_height заменим медианным занчением.
median_ceiling_height=data['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(median_ceiling_height)
#Пропуски в данных living_area и kitchen_area заменяем на -1
data['living_area'] = data['living_area'].fillna(-1)
data['kitchen_area'] = data['kitchen_area'].fillna(-1)
#Попуски в is_apartment заменяем False
data['is_apartment'] = data['is_apartment'].fillna(False)
#Пропуски из days_exposition заменяем на -1
data['days_exposition'] = data['days_exposition'].fillna(-1)
#Данные из balcony заполним пропуски 0
data['balcony'] = data['balcony'].fillna(0)
#Пропуски из parks_around_3000  заполнить 0
data['parks_around_3000'] = data['parks_around_3000'].fillna(0)
#Пропуски из ponds_around_3000 нужно заполнить 0
data['ponds_around_3000'] = data['ponds_around_3000'].fillna(0)
#airports_nearest, сity_centers_nearest, parks_nearest, ponds_nearest заполняем -1
data['airports_nearest'] = data['airports_nearest'].fillna(-1)
data['сity_centers_nearest'] = data['сity_centers_nearest'].fillna(-1)
data['ponds_nearest'] = data['ponds_nearest'].fillna(-1)
data['parks_nearest'] = data['parks_nearest'].fillna(-1)
```

##### Удаление пропусков

```python
data.dropna(subset=['floors_total'], inplace=True)
```

#### Изменяем типы данных

```python
# Для удобства поменяем тип данных в last_price на int64
data['last_price'] = data['last_price'].astype(int)
#Данные в колонке first_day_exposition представлены в формате object, должны быть в datetime64. 
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')
#Данные из days_exposition нужно перевести из float64 в int64 
data['days_exposition'] = data['days_exposition'].astype(int)
#Данные из balcony нужно перевести из float64 в int64 
data['balcony'] = data['balcony'].astype(int)
#Данные из parks_around_3000 нужно перевести из float64 в int64
data['parks_around_3000'] = data['parks_around_3000'].astype(int)
#Данные из ponds_around_3000 нужно перевести из float64 в int64 
data['ponds_around_3000'] = data['ponds_around_3000'].astype(int)
#Данные из floors_total нужно перевести из float64 в int64 
data['floors_total'] = data['floors_total'].astype(int)
#Данные из is_apartment нужно перевести из float64 в int64
data['is_apartment'] = data['is_apartment'].astype(bool)
data['days_exposition'] = data['days_exposition'].astype(int)                      
data['balcony'] = data['balcony'].astype(int)                                                  
#Данные из parks_around_3000 нужно перевести из float64 в int64
data['parks_around_3000'] = data['parks_around_3000'].astype(int)
#Данные из ponds_around_3000 нужно перевести из float64 в int64
data['ponds_around_3000'] = data['ponds_around_3000'].astype(int)

```

#### Посчитайте среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений

Посчитайте среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений. Выделите населённые пункты с самой высокой и низкой стоимостью квадратного метра. Эти данные можно найти по имени в столбце locality_name.


```python
top_10=data['locality_name'].value_counts().head(10)
top_10_price=data.query('locality_name in (@top_10.index)') \
.pivot_table(index='locality_name', values='square_meter_price', aggfunc='mean') \
.sort_values(by='square_meter_price', ascending=False)
top_10_price = top_10_price.rename(columns={'square_meter_price':'цена за квадратный метр'}) 
display(top_10_price)
```

Цена квадратный метр самая большая в Санкт-Петербурге.
